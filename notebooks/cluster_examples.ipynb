<h2>Clustering Algorithms Examples</h2>

```python
from matplotlib import pyplot as plt
from src.workflow import Workflow
from utils.utils_experiment import load_data
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = 'spleen'
X, Y = load_data(dataset)
w = Workflow(X, Y, config=dataset)
```

```python
w.reduce_dim(method='pca')
```

<h3>k-Means Clustering</h3>

```python
w.cluster(method='kmeans')
w.reduce_plot(labels=w.y_train_pred, method='umap')
```

<h3>Spectral Clustering</h3>

```python
w.cluster(method='spectral')
w.reduce_plot(labels=w.y_train_pred, method='umap')
```

<h3>k-Medoids Clustering</h3>

```python
w.cluster(method='kmedoids')
w.reduce_plot(labels=w.y_train_pred, method='umap')
```
